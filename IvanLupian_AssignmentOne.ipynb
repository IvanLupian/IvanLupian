## 4. Cleaning of Database

```python
print(data.isnull().sum())
```

```python
print(data.duplicated())
print('Number of duplicates: ', data.duplicated().sum())
```

```python
noNaN_data = data.dropna()
clean_data = noNaN_data.drop_duplicates()
print('Previous data shape: ', data.shape)
print('After cleaning data shape: ',clean_data.shape)
x = data.shape[0]
y = clean_data.shape[0]
red_proc = ((x-y)/x)*100
print('Reduced by: ', format(red_proc, '.2f'), '% from the original database')
print('Number of duplicates: ', clean_data.duplicated().sum())
print('Number of missing values:\n',clean_data.isnull().sum())
```

## 5. Variables of interest

```python
variables = clean_data[["maker", "runned_miles", "adv_year", "reg_year", "top_speed_mph", "fuel_type", "price"]]
variables
```

## 6. Boxplot to identify outliers

```python
plt.figure(figsize=(15,8))
sns.boxplot(x=variables['reg_year'],
            y=variables['price'])
plt.xlabel('Registration year')
plt.xticks(rotation=45)
plt.ylabel('Price of the car')
plt.title('Car Value over Time')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(x=variables['fuel_type'],
            y=variables['top_speed_mph'])
plt.xlabel('Fuel type')
plt.xticks(rotation=45)
plt.ylabel('Max Speed (mph)')
plt.title('Maximum Speed over Fuel Type')
plt.show()
```

## 7. IQM to remove outliers

```python
Q1 = variables['price'].quantile(0.25)
Q3 = variables['price'].quantile(0.75)
IQR = Q3-Q1

upper_bound = Q3 + 1.5*IQR
lower_bound = Q1 - 1.5*IQR

cleaned_data = variables[(variables['price'] >lower_bound)&(variables['price']<upper_bound)]
plt.figure(figsize=(15,8))
sns.boxplot(x=cleaned_data['reg_year'].astype('Int64'),
            y=cleaned_data['price'])
plt.xlabel('Registration year')
plt.ylabel('Price of the car')
plt.xticks(rotation = 45)
plt.title('Car Value over Time -- Clean')
plt.show()

print('Maximum possible value: ', upper_bound)
print('Minimum possible value: ', lower_bound)
```

```python
Q1 = variables['top_speed_mph'].quantile(0.25)*1.60934
Q3 = variables['top_speed_mph'].quantile(0.75)*1.60934
IQR = Q3-Q1

upper_bound = Q3 + 1.5*IQR
upper_bound = round(upper_bound, 2)
lower_bound = Q1 - 1.5*IQR
lower_bound = round(lower_bound, 2)

cleaned_data = variables[(variables['top_speed_mph'] >lower_bound)&(variables['top_speed_mph']<upper_bound)]
plt.figure(figsize=(10,8))
sns.boxplot(x=cleaned_data['fuel_type'],
            y=cleaned_data['top_speed_mph'])
plt.xlabel('Fuel type')
plt.ylabel('Max Speed (km/h)')
plt.xticks(rotation = 45)
plt.title('Maximum Speed over Fuel Type -- Clean')
plt.show()

print('Maximum possible value: ', upper_bound)
print('Minimum possible value: ', lower_bound)
```

###  Manual Method to remove outliers

```python
cleaned_data = variables[(variables['top_speed_mph'] >120)&(variables['top_speed_mph']<195)]
plt.figure(figsize=(10,8))
sns.boxplot(x=cleaned_data['fuel_type'],
            y=cleaned_data['top_speed_mph'])
plt.xlabel('Fuel type')
plt.ylabel('Max Speed (km/h)')
plt.xticks(rotation = 45)
plt.title('Maximum Speed over Fuel Type -- Clean')
plt.show()
```
